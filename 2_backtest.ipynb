# 歷史回測的限制
    1. 歷史回測相信「歷史會不斷重現」，過去績效好不代表未來績效也會好
    2. 無法捕捉交互作用，例如：market impact，想想看，如果你今天買進4萬張台積電，會不會影響台積電價格？
